```python
customer_df = customer_df.set_index('Customer')
```

```python
# 4 Load the continuous and discrete variables into numericals_df and categorical_df variables, for eg.:
#  numerical_df = customer_df.select_dtypes()
#  categorical_df = customer_df.select_dtypes()
```

```python
numerical_df = customer_df.select_dtypes('int64','float64')
categorical_df = customer_df.select_dtypes('object')
```

```python
# 5 Plot every categorical variable. What can you see in the plots? Note that in the previous lab you used a bar plot to plot 
# categorical data, with each unique category in the column on the x-axis and an appropriate measure on the y-axis. However,
# this time you will try a different plot. This time in each plot for the categorical variable you will have, each unique
#category in the column on the x-axis and the target(which is numerical) on the Y-axis
```

```python
for col in categorical_df:
    sns.countplot(x=col, data=customer_df)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 6 For the categorical data, check if there is any data cleaning that need to perform. Hint: You can use the function
# value_counts() on each of the categorical columns and check the representation of different categories in each column. 
# Discuss if this information might in some way be used for data cleaning.
```

```python
for col in customer_df.select_dtypes(include="object").columns:
    value_counts = customer_df[col].value_counts()

    print(f"Columna: {col}")
    print(value_counts)
    print("\n")
```
