```python
drop_columns=['pvastate','recp3','veterans']
data = data.drop(drop_columns, axis=1)
```

```python
data
```

```python
#domain
data.domain.value_counts()
```

```python
data = data[data['domain'] != " "]

data = data.reset_index()
data.domain.value_counts()
```

```python
data.shape
```

```python
#gender
data.gender.value_counts()
```

```python
def clean_gender(x):
    if x in [' ','U','J','C','A']:
        return 'other'
    else:
        return x
```

```python
data['gender'] = list(map(clean_gender,data['gender']))
data.gender.value_counts()
```

```python
sns.boxplot(x='gender', y='avggift', data=data)
plt.show()
```

```python
sns.barplot(x='gender', y='avggift', data=data)
plt.show()
```

```python
data = data.drop(['gender'], axis=1)
data
```

```python
#state
data.state.nunique()
```

```python
vals = data.state.value_counts()
vals
```

```python
vals = pd.DataFrame(vals)
vals = vals.reset_index()
```

```python
vals
```

```python
vals.columns = ['state', 'counts']
#vals.columns=['counts']
vals
```

```python
group_states = vals[vals['counts']<2590]
group_states
```

```python
group_states = list(group_states['state'])
group_states
```

```python
def clean_state(x):
    if x in group_states:
        return 'other'
    else:
        return x
```

```python
data['state'] = list(map(clean_state, data['state']))
data['state'].value_counts()
```

Bucketing : [pd.cut()](https://pandas.pydata.org/docs/reference/api/pandas.cut.html) and [pd.qcut()](https://pandas.pydata.org/docs/reference/api/pandas.qcut.html)

```python
data.ic2.nunique()
```

```python
data.ic2.describe()
```

```python
ic2_labels = ['Low', 'Moderate', 'High', 'Very High']
data['ic2_'] = pd.cut(data['ic2'],4,labels=ic2_labels)
data
```

```python
data.ic2_.dtype
```

```python
data.ic2_
```

```python
ic2_q = pd.qcut(data.ic2,4)
ic2_q
```

```python
sns.barplot(x='ic2_', y='avggift', data=data)
plt.show()
```
